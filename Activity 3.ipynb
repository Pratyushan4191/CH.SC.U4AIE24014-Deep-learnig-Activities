```python
import numpy as np
import pandas as pd
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, log_loss

# XOR Dataset
X = np.array([
    [0,0],
    [0,1],
    [1,0],
    [1,1]
])

y = np.array([0,1,1,0])

# Train-validation split
X_train, X_val, y_train, y_val = train_test_split(
    X, y,
    test_size=0.25,
    random_state=42
)

# Hyperparameter combinations
configs = [
    {"lr":0.01,"hidden":4},
    {"lr":0.1,"hidden":4},
    {"lr":0.01,"hidden":8},
    {"lr":0.1,"hidden":8},
    {"lr":0.05,"hidden":6}
]

results = []

for config in configs:

    model = MLPClassifier(
        hidden_layer_sizes=(config["hidden"],),
        activation='logistic',
        learning_rate_init=config["lr"],
        max_iter=1000,
        random_state=42
    )

    model.fit(X_train,y_train)

    train_prob = model.predict_proba(X_train)
    val_pred = model.predict(X_val)

    train_loss = log_loss(y_train,train_prob)
    val_acc = accuracy_score(y_val,val_pred)

    results.append([
        config["lr"],
        config["hidden"],
        round(train_loss,4),
        round(val_acc*100,2)
    ])

results_df = pd.DataFrame(
    results,
    columns=[
        "Learning Rate",
        "Hidden Units",
        "Training Loss",
        "Validation Accuracy (%)"
    ]
)

print(results_df)
```

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score
)

val_pred = model.predict(X_val)

acc = accuracy_score(y_val, val_pred)

prec = precision_score(
    y_val,
    val_pred,
    zero_division=0
)

rec = recall_score(
    y_val,
    val_pred,
    zero_division=0
)

print("Accuracy :", acc)
print("Precision:", prec)
print("Recall   :", rec)
```
